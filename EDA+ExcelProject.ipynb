```python
import matplotlib.pyplot as plt

job_counts = df2['JobTitle'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(job_counts.index, job_counts.values)
plt.title('Count Plot of Top 10 Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
job_counts = df2['JobTitle'].value_counts().head(5)

plt.figure(figsize=(6,6))
plt.pie(job_counts.values, labels=job_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart of Top 5 Job Titles')
plt.show()
```
